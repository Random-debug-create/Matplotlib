# 1.Line Graphs using Matplotlib and Numpy

```python
#Creating an array
a = np.array([1,2,3,4,5])
a

plt.plot(a)
plt.show()
```

```python
# creating 2 arrays

a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])

plt.plot(a)
plt.plot(b)
plt.show()
```

```python
# Having more than 1 line - Adding Title, Legends and x/y tickers

#creating the row
a = np.array([1,2,3,4,5])
b = np.array([2,2,2,2,2])
c = np.array([3,2,4,6,5])

# Increase size 
plt.figure(figsize=(15,10))

plt.plot(a, color = "green", linestyle = "-.", label="Rent FY20")
plt.plot(b, linewidth=5., label="AVG Rent")
plt.plot(c, color = "red", linestyle = "--", label="Rent FY19")
plt.title("Rent per Month Summary", fontsize=20, fontweight="bold")
plt.legend(loc="upper left", fontsize=15)
plt.xticks(range(0,len(a)),("Jan","Feb","Mar", "Apr","May"))
plt.xlabel("Months")
plt.ylabel("Rent $")
plt.show()
```

# 2. Scatter Plots using Matplotlib and Numpy

```python
# np.arange: from 0 to the number you specify
# np.random.randit: From -> To ,Number of numbers you need
# np.random.randn(): random numbers with normal distribution (mena 0 and variance 1)

a = np.arange(30)
b = np.random.randint(0,30,30)

plt.scatter(a,b)
plt.show()
```

```python
#Scatter with size

s = np.random.randint(0,100,30)
b = a + np.random.randint(0,30,30)  #a on juhuslik arvude massiiv. b saadakse a peale liites veel ühed juhuslikud arvud vahemikus 0 kuni 29. See tähendab, et iga b väärtus on vähemalt a väärtus, aga võib olla kuni 29 võrra suurem. Seega b on alati võrdne või suurem kui a, aga mitte väiksem.

plt.scatter(a,b, c="green", s=s)    #s= - Näitab suurust, s - s = np.random.randint(0,100,30)
plt.show()
```

# 3. Bar Plots using Matplotlib and Numpy

```python
#Creating the raw data
category = ["Account A","Account B", "Account C"]
revenue = np.array([300, 500, 700])

# Plotting it
plt.figure(figsize=(10,10))
plt.bar(range(0,3), revenue)
plt.xticks(range(0,3), category)
plt.show()
```

```python
# Subplots

# Creating the raw data
category = ["Account A", "Account B", "Account C"]
revenue = np.array([300, 450, 700])

# Plotting it 
plt.figure(figsize=(10,5))

# Bar
plt.subplot(131)  # 131 - Tabeli järjekorra number reas
plt.bar(range(0,3), revenue)
plt.xticks(range(0,3), category)

# Line
plt.subplot(132) # 132 - Tabeli järjekorra number reas
plt.plot(range(0,3), revenue, c="red")
plt.xticks(range(0,3), category)

# Scatter
plt.subplot(133) # 133 - Tabeli järjekorra number reas
plt.scatter(range(0,3), revenue, c="green")
plt.xticks(range(0,3), category)

plt.show()
```

```python
# Loading the data

raw_data = pd.read_csv("C:\\Users\\User\\OneDrive\\Töölaud\\Töö\\Port\\Pyhton\\Matplotlib\\Marketing Raw Data.csv")
#Runs all the data
raw_data

print(raw_data.shape)

#Runs the first 5 rows
raw_data.head()
```

```python
# Line graph

plt.plot(raw_data["Visitors"])
plt.show()
```

```python
# Line graph with xtickers

plt.plot(raw_data["Visitors"])
plt.xticks(range(0,len(raw_data["Visitors"])), raw_data["Date"], rotation = 90)  #Kuupäevad x -teljel on tihedalt koos kuna ruumi nende jaoks on vähe
plt.show()
```

```python
# Line graphs with xtickers and less data

a = raw_data.tail(10).reset_index() #Viimased 10 andmepunkti

plt.plot(a["Visitors"])
plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.show()
```

```python
# 2 data points on the graph

plt.figure(figsize=(15,10))
plt.plot(a["Visitors"], label="N0. of Visitors", linewidth=5.)
plt.plot(a["Revenue"], linestyle = "-.", label="Revenue")
plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors and Revenue by Day", fontsize=20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Dates")
plt.show()
```

```python
# Combo graphs

plt.figure(figsize=(15,10))
plt.bar(range(0,len(a["Visitors"])), a["Visitors"], label="No. of Visitors", linewidth=5, color = "green")
plt.plot(a["Revenue"], linestyle = "-.", label="Revenue")
plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors and Revenue by Day" , fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Dates")
plt.show()
```

```python
# Combo stack bars

plt.figure(figsize=(15,10))
plt.bar(range(0,len(a["Revenue"])), a["Revenue"], label="Revenue", linewidth=5, color = "green")
plt.bar(range(0,len(a["Visitors"])), a["Marketing Spend"], label="Marketing Spend", linewidth=5, color = "red")
plt.plot(a["Visitors"], linestyle = "-.", label="Revenue")
plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors and Revenue by Day" , fontsize = 20)
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Dates")
plt.show()
```

```python
# Combo 2+ bargroups + secondary scale

weight = .30

a[["Revenue", "Marketing Spend"]].plot(kind="bar", color = ["green", "red"], figsize=(15,10))
a["Visitors"].plot(c="blue", linestyle = "-.", secondary_y = True)

plt.xticks(range(0,len(a["Visitors"])), a["Date"], rotation = 90)
plt.title("Visitors and Revenue by Day", fontsize = 20) 
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Dates")
plt.xlim(-weight, len(a["Visitors"])-weight)
plt.show()
```

# 6. Scatterplots Using DFs


```python
# Scatterplot

plt.figure(figsize=(15,10))
plt.scatter(raw_data["Revenue"], raw_data["Marketing Spend"], c="red", marker="+", label="Revenue")
plt.title("Relationship between M. Spend and Revenue", fontsize=20)
plt.legend(loc="upper left", fontsize=14)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```

```python
raw_data.head()
```

```python
# Scatterplot with groups

groups = raw_data[["Promo","Revenue","Marketing Spend"]].groupby("Promo")
groups

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group["Revenue"], group["Marketing Spend"], marker = "x", label=promo)

ax.legend()
plt.title("Relationship between M. Spend andRevenue by Promo", fontsize = 20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```

```python
groups = raw_data[["Day_Name","Revenue","Marketing Spend"]].groupby("Day_Name")
groups

fig, ax = plt.subplots(figsize=(15,10))

for promo, group in groups:
    ax.scatter(group["Revenue"], group["Marketing Spend"], marker = "x", label=promo)

ax.legend()
plt.title("Relationship between M. Spend andRevenue by Day_Name", fontsize = 20)
plt.xlabel("Revenue")
plt.ylabel("Marketing Spend")

plt.show()
```
